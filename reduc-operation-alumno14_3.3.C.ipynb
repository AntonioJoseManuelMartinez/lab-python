```python
import time
import sys

def operacion_reduccion(a):
    """Suma los números del rango [0, a)."""
    suma = 0
    for i in range(a):
        suma += i
    return suma

# Determinar value según si se ejecuta desde Jupyter o desde sbatch
if len(sys.argv) > 1:
    try:
        value = int(sys.argv[1])
        print("\nPara un valor", value, "\n")
    except:
        print("Argumento no válido. Usando valor por defecto 10**6.")
        value = 10**6
else:
    value = 10**6
    print("Modo Jupyter: usando valor por defecto:", value, "\n")

# Medir tiempo de ejecución
t_inicial = time.time()
resultado = operacion_reduccion(value)
t_final = time.time()

print("Time taken by reduction operation:", t_final - t_inicial, "seconds\n")
print("\tComputing the sum of numbers in the range [0, value):", resultado, "\n")
```
